```python
import gensim
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import re
```

```python
corpus = pd.read_csv("cleaned_data.csv")
```

```python
corpus.drop("Unnamed: 0", axis = 1, inplace = True)
corpus.dropna(subset = ["text"], inplace = True)
display(corpus)
```

```python
ps = PorterStemmer()
```

```python
def stemmer(text):
    words = word_tokenize(text)
    stemmed_value = ""
    for word in words:
        stemmed_value += ps.stem(word)+" "
    return stemmed_value
```

```python
corpus ["stemmed_text"] = corpus ["text"].apply(lambda x: stemmer(x))
```

```python
liwc_dictionary = {}
```

```python
counter = 0
with open('LIWC2001_English.dic') as f:
    while True:
        line = f.readline()
        if not line:
            break
        if counter == 0:
            pass
        else:
            line_striped = line.strip()
            line_striped = line_striped.split()
            l1 = []
            for i in range(1, len(line_striped)):
                l1.append(int(line_striped[i]))                 
            liwc_dictionary [line_striped [0]] = l1
            
        counter += 1
```

```python
def calc_first_person_pronouns(text):
    words = text.split(" ")
    self_prn_counter = 0
    for word in words:
        for key in liwc_dictionary.keys():
            temp = re.compile(key)
            result = temp.match(word)
            if result != None:
                if 4 in liwc_dictionary [key]:
                    self_prn_counter += 1
    return self_prn_counter
```
